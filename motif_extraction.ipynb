```python

from Bio import SeqIO

def return_sequence_from_fasta(fasta_location):
    full_fasta = list(SeqIO.parse(fasta_location,"fasta"))
    sequence = (full_fasta[0].seq)
    return str(sequence)
```

```python
full_unicorn_sequence = return_sequence_from_fasta("sequence.fasta")
```

```python
def write_to_txt (string, filename):
    with open(filename, "a") as f:
        f.write(string + "\n")
def read_txt_and_return_lines(filename):
    with open(filename, "r") as f:
        lines = [line.strip() for line in f]
    return lines
```

```python
TXT_NAME_FULL_DNA = "full_dna.txt"
write_to_txt(full_unicorn_sequence, TXT_NAME_FULL_DNA)
```

```python
import random

def extract_motif(sequence, length):
    CODON_LEN = 3
    start = random.randrange(0, len(sequence) - length, CODON_LEN)
    return sequence[start:start + length]
SEQ_LEN = [9, 12, 15, 18]
```

```python
random_seq_dict = {}
for seq_len in SEQ_LEN:
    random_seq_dict[seq_len] = extract_motif(full_unicorn_sequence, seq_len)
```

```python
def embed_in_random_seq(motif, factor):
    total_length = len(motif) * factor
    background = ''.join(random.choices('ATGC', k=total_length - len(motif)))
    insert_pos = random.randint(0, len(background))
    return background[:insert_pos] + motif + background[insert_pos:]
```

```python
FACTOR = 10
ten_fold_random_dict = {}
for seq_len in SEQ_LEN:
    ten_fold_random_dict[seq_len] = embed_in_random_seq(random_seq_dict[seq_len], FACTOR)
ten_fold_random_dict["full_seq"] = embed_in_random_seq(full_unicorn_sequence, FACTOR)
```

```python
def random_list_generator(motif, amount, factor):
    return [embed_in_random_seq(motif, factor) for _ in range(amount)]
    
def dictionary_values_generator(dict_input, amount, factor):
    random_dict = {}
    for key, value in dict_input.items():
        random_dict[key] = random_list_generator(value, amount, factor)
    return random_dict
```

```python
NUMBER_OF_SEQUENCES = 20
twenty_random_seq_dict = dictionary_values_generator(random_seq_dict, NUMBER_OF_SEQUENCES, FACTOR)
for key in twenty_random_seq_dict.keys():
    FILE_NAME = str(key) + "_char_seq.txt"
    for seq in twenty_random_seq_dict[key]:
        write_to_txt(seq, FILE_NAME)
    
```

```python
full_len_random_seq_list = random_list_generator(full_unicorn_sequence, NUMBER_OF_SEQUENCES, FACTOR)
for seq in full_len_random_seq_list:
    write_to_txt(seq, "full_char_seq.txt")
```

```python
MOTIFS_FILE_NAME = "motifs.txt"
for key,value in random_seq_dict.items():
    input_string = f"{key}-long sequence motif: {value}"
    write_to_txt(input_string, MOTIFS_FILE_NAME)
```
